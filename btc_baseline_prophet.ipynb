```python
import prophet
import pandas as pd
import yfinance as yf
```

```python
data = yf.download(['BTC-USD'])['2020':]
```

```python
data['y'] = data['Close']
data['ds'] = data.index
```

```python
data
```

```python
model = prophet.Prophet(daily_seasonality = True, weekly_seasonality = True)
model.fit(data[['ds', 'y']])
```
